# Statistical Learning

### What Is Statistical Learning?

```python
adv = pd.read_csv('data/Advertising.csv')
adv.head(2)
```

**Input Variable or Predictor ($X$)**: Advertising budget 

**Output Variable or Response ($Y$)**: Sales

```python
fig, axs = plt.subplots(ncols=3, figsize=(12,4))
sns.regplot(x="TV",y="sales", data=adv, ax=axs[0],scatter_kws={"color": "red"})
sns.regplot(x="radio",y="sales", data=adv, ax=axs[1],scatter_kws={"color": "red"})
sns.regplot(x="newspaper",y="sales", data=adv, ax=axs[2],scatter_kws={"color": "red"})

fig.suptitle('FIGURE 2.1. The Advertising data set',
             fontsize=18)
```

We assume a relationship between $Y$ and $X = (X_1,X_2,...,X_n$), which has the general form
$$Y = f(X) + \epsilon$$

```python
income = pd.read_csv('data/Income2.csv')
income.head(2)
```

```python
fig, axs = plt.subplots(ncols=2, figsize=(12,6))
fig.suptitle('FIGURE 2.2 The Income data set', fontsize=18)
sns.regplot("Education", "Income",
           data=income,fit_reg=False, ax=axs[0],scatter_kws={"color": "red"})
sns.regplot("Education", "Income",
           data=income,fit_reg=True, order=3, ax=axs[1],scatter_kws={"color": "red"})
plt.xlabel('Years of Education')
```

The blue line represents the relation ($f$) between Income and Education, the observations that lies above the curve shows *positive* errors and those that lies below shows *negative* erros ($\epsilon$). In Statistical Learning we always want to estimate $f$.

## Why Estimate $f$

**Prediction**

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

xs = income['Income']
ys = income['Education']
zs = income['Seniority']
ax.scatter(xs, ys, zs, s=50, alpha=0.6,
           edgecolors='w', c='r')

ax.set_xlabel('Income')
ax.set_ylabel('Years of Education')
ax.set_zlabel('Seniority')

plt.show()
```
